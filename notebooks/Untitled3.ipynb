```python
import os
import pandas as pd
from sqlalchemy import create_engine
from urllib.parse import quote_plus
from dotenv import load_dotenv

load_dotenv()  # reads .env in the current folder

engine = create_engine(
    f"postgresql+psycopg2://{os.getenv('PGUSER')}:{quote_plus(os.getenv('PGPASSWORD'))}"
    f"@{os.getenv('PGHOST')}:{os.getenv('PGPORT')}/{os.getenv('PGDATABASE')}"
)
```

```python
pd.read_sql("SELECT current_database() AS db, current_user AS usr, now() AS ts;", engine)
```

```python
with engine.begin() as conn:
    conn.exec_driver_sql("""
        CREATE TABLE IF NOT EXISTS public.python_check (
            id serial PRIMARY KEY,
            created_at timestamptz DEFAULT now()
        );
    """)
    conn.exec_driver_sql("INSERT INTO public.python_check DEFAULT VALUES;")

pd.read_sql("SELECT * FROM public.python_check ORDER BY id DESC LIMIT 5;", engine)
```
