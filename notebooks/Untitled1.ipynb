```python
import pandas as pd
from sqlalchemy import create_engine
from urllib.parse import quote_plus

user = "postgres"
password = "SamMsh!@25"
host = "localhost"
port = 5432
db = "analytics_db"   # <-- switch here

engine = create_engine(
    f"postgresql+psycopg2://{user}:{quote_plus(password)}@{host}:{port}/{db}"
)

pd.read_sql("SELECT current_database() as db, now() as connected_time;", engine)
```
